```python
categorical_cols = ['gender', 'academic_level', 'internet_quality', 'part_time_job']
```

```python
for col in categorical_cols:
    print(f"Value counts for {col}: \n {df[col].value_counts()}")
```

```python
df.hist(bins=20, edgecolor="black")
plt.tight_layout()
plt.show()
```

```python
for col in categorical_cols:
  sns.countplot(data=df, x=col)
  plt.title(f"Distribution of {col}")
  plt.xticks(rotation=45)
  plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
df.describe().columns
```

```python
num_features = ['student_id', 'age', 'study_hours', 'self_study_hours',
       'online_classes_hours', 'social_media_hours', 'gaming_hours',
       'sleep_hours', 'screen_time_hours', 'exercise_minutes',
       'caffeine_intake_mg', 'part_time_job', 'upcoming_deadline',
       'mental_health_score', 'focus_index', 'burnout_level',
       'productivity_score']
```

```python
for features in num_features:
  sns.scatterplot(data = df, x=features, y="exam_score")
  plt.title(f"{features} vs Exam Score")
  plt.show()
```

```python
for col in categorical_cols:
  sns.boxplot(data = df, x=col, y="exam_score")
  plt.title(f"Exam Score by {col}")
  plt.xticks(rotation=45)
  plt.show()
```
